## Iterate through folders

```python
str_train = ['MediumResidential','Park','Parking','School','Square']
num_train = [290,350,390,300,330]
str_test = ['DenseResidential','SparseResidential'] #Blindtesting
num_test = [410,300]
str_mult = ['x1','x2','x4']
```

```python
base_dir = '../../Data/AID/'
for i in range(len(str_train)):
  train = str_train[i]
  ind_x1 = []
  ind_x2 = []
  ind_x4 = []
  for j in range(1,(num_train[i]+1)):
    if not os.path.exists(base_dir + 'AIDx1/' + train + '/' + train.lower() + '_' + str(j) + '.png'):
      ind_x1.append(str(j))
    if not os.path.exists(base_dir + 'AIDx2/' + train + '/' + train.lower() + '_' + str(j) + '.png'):
      ind_x2.append(str(j))
    if not os.path.exists(base_dir + 'AIDx4/' + train + '/' + train.lower() + '_' + str(j) + '.png'):
      ind_x4.append(str(j))
  print(train)
  print('x1: ' + str(ind_x1))  
  print('x2: ' + str(ind_x2))
  print('x4: ' + str(ind_x4))
```

```python
base_dir = '../../Data/AID/'
for i in range(len(str_test)):
  test = str_test[i]
  ind_x1 = []
  ind_x2 = []
  ind_x4 = []
  for j in range(1,(num_test[i]+1)):
    if not os.path.exists(base_dir + 'AIDx1/' + test + '/' + test.lower() + '_' + str(j) + '.png'):
      ind_x1.append(str(j))
    if not os.path.exists(base_dir + 'AIDx2/' + test + '/' + test.lower() + '_' + str(j) + '.png'):
      ind_x2.append(str(j))
    if not os.path.exists(base_dir + 'AIDx4/' + test + '/' + test.lower() + '_' + str(j) + '.png'):
      ind_x4.append(str(j))
  print(test)
  print('x1: ' + str(ind_x1))  
  print('x2: ' + str(ind_x2))
  print('x4: ' + str(ind_x4))
```
