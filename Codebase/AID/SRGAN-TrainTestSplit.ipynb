# Imports

```python
import numpy as np
import os

from pathlib import Path

from PIL import Image
 
import shutil 
import re
```

## Extracting from folders

```python
dir_trainAID = "../../Data/AID/Train1_5class/"
str_train = ['train_HR.csv','train_LRx2.csv','train_LRx4.csv','valid_HR.csv','valid_LRx2.csv','valid_LRx4.csv']
dir_testAID = "../../Data/AID/Test1_2class/"
str_test = ['test_HR.csv','test_LRx2.csv','test_LRx4.csv']
```

```python
dir_train_HR = "../../Data/AID/Train1_5class/train_HR/"
dir_train_LRx2 = "../../Data/AID/Train1_5class/train_LRx2/"
dir_train_LRx4 = "../../Data/AID/Train1_5class/train_LRx4/"
dir_valid_HR = "../../Data/AID/Train1_5class/valid_HR/"
dir_valid_LRx2 = "../../Data/AID/Train1_5class/valid_LRx2/"
dir_valid_LRx4 = "../../Data/AID/Train1_5class/valid_LRx4/"
dir_test_HR = "../../Data/AID/Test1_2class/test_HR/"
dir_test_LRx2 = "../../Data/AID/Test1_2class/test_LRx2/"
dir_test_LRx4 = "../../Data/AID/Test1_2class/test_LRx4/"

Path(dir_train_HR).mkdir(parents=True, exist_ok=True)
Path(dir_train_LRx2).mkdir(parents=True, exist_ok=True)
Path(dir_train_LRx4).mkdir(parents=True, exist_ok=True)
Path(dir_valid_HR).mkdir(parents=True, exist_ok=True)
Path(dir_valid_LRx2).mkdir(parents=True, exist_ok=True)
Path(dir_valid_LRx4).mkdir(parents=True, exist_ok=True)
Path(dir_test_HR).mkdir(parents=True, exist_ok=True)
Path(dir_test_LRx2).mkdir(parents=True, exist_ok=True)
Path(dir_test_LRx4).mkdir(parents=True, exist_ok=True)
```

```python
for train in str_train:
  print(train)
  files = np.genfromtxt(dir_trainAID + train,dtype='str')
  for file in files:    
    ind = [m.start() for m in list(re.finditer(r'/',file))]
    shutil.copyfile(file, dir_trainAID + train[:-4] + '/' + file[ind[-1]+1:])
```

```python
for test in str_test:
  print(test)
  files = np.genfromtxt(dir_testAID + test,dtype='str')
  for file in files:    
    ind = [m.start() for m in list(re.finditer(r'/',file))]
    shutil.copyfile(file, dir_testAID + test[:-4] + '/' + file[ind[-1]+1:])
```

## Verify files

```python
for train in str_train:
  list_files = [] 
  print(train)
  for root, dirs, files in os.walk(dir_trainAID + train[:-4]):
    for name in files:
        list_files.append(os.path.join(root, name))
  print(len(list_files))
  print(list_files[:5])
```
