```python
files_train_LRx8 = []
files_valid_LRx8 = []
files_test_LRx8 = []

with open(file_train, newline='') as csvfile:
    files_train = csv.reader(csvfile, delimiter=' ')
    for file in files_train:
      ind = [m.start() for m in list(re.finditer(r'/',file[0]))]
      files_train_LRx8.append(file[0][:ind[-2]-1] + '8' + file[0][ind[-2]:])
      
with open(file_valid, newline='') as csvfile:
    files_valid = csv.reader(csvfile, delimiter=' ')
    for file in files_valid:
      ind = [m.start() for m in list(re.finditer(r'/',file[0]))]
      files_valid_LRx8.append(file[0][:ind[-2]-1] + '8' + file[0][ind[-2]:])
      
with open(file_test, newline='') as csvfile:
    files_test = csv.reader(csvfile, delimiter=' ')
    for file in files_test:
      ind = [m.start() for m in list(re.finditer(r'/',file[0]))]
      files_test_LRx8.append(file[0][:ind[-2]-1] + '8' + file[0][ind[-2]:])
```
