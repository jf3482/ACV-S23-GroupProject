# Imports

```python
import numpy as np
import os

from pathlib import Path

from PIL import Image

from sklearn.model_selection import train_test_split

import shutil
import re
```

## Extracting from folders

```python
dir_data = '../../Data/AID/AIDoriginal'
```

```python
list_folders = []
list_files = []
for root, dirs, files in os.walk(dir_data):
   for name in files:
      list_files.append(os.path.join(root, name))
   for name in dirs:
      list_folders.append(os.path.join(root, name))
```

# Downsample images via Pillow (bicubic)

```python
#Added additional x6 case (post extraction)
dir_x6data = "../../Data/AID/AIDx6"
for folder in list_folders:
  Path(dir_x6data + "//" + folder[27:]).mkdir(parents=True, exist_ok=True)

#Initialization of parameters
width, height = 600,600

# Run once
#Going through data and downsampling
for file in list_files:
    # Opens a image in RGB mode
    im = Image.open(file)
 
    newsize3 = (int(width/6), int(height/6))
    imx6 = im.resize(newsize3)
    imx6.save(dir_x6data + "/" + file[27:-4] + '.png')
    
```
